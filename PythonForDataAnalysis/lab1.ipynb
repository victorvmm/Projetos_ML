# IMPORTS E LEITURA DE TABELA

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('gapminder.csv')
data
```

# REMOÇÃO DE COLUNAS IRRELEVANTES

```python
data = data.drop(columns=["polityscore", "urbanrate", "femaleemployrate", "breastcancerper100th", 'co2emissions', 'oilperperson', 'relectricperperson', 'hivrate', 'armedforcesrate', 'internetuserate'])
data
```

```python
data.info()
```

# CONVERSÃO PARA TIPOS ADEQUADOS DE DADOS E REMOÇÃO DAS LINHAS COM NULO (OUTLIERS)

```python
data = data.convert_dtypes()
data.info()
data
```

```python
data = data.replace(np.nan, '0')
data
```

```python
data['incomeperperson'] = round(data['incomeperperson'].astype(float), 2)
data['alcconsumption'] = round(data['alcconsumption'].astype(float), 2)
data['lifeexpectancy'] = round(data['lifeexpectancy'].astype(float), 0)
data['lifeexpectancy'] = data['lifeexpectancy'].astype(int)
data['suicideper100th'] = round(data['suicideper100th'].astype(float), 2)
data['employrate'] = round(data['employrate'].astype(float), 2)
data
```

```python
data.info()
```

```python
data = data.replace(0, np.nan)
data = data.dropna()
data
```
